```python
df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')

df['discount_percentage'] = df['discount_percentage'] / 100
```

```python
df['rating'].value_counts()
```

```python
df.query('rating == "|"')
```

```python
df['rating'] = df['rating'].astype(str).str.replace('|', '3.9').astype('float64')
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '').astype('float64')
```

```python
plt.figure(figsize=(22, 10))
# plot the null values in each column
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
plt.figure(figsize=(22, 10))
# plot the null values by their percentage in each column
missing_percentage = df.isnull().sum()/len(df)*100
missing_percentage.plot(kind='bar')
# add the labels
plt.xlabel('Columns')
plt.ylabel('Percentage')
plt.title('Percentage of Missing Values in each Column')
```

```python
df[df['rating_count'].isnull()].head(5)
```

```python
df['rating_count'] = df.rating_count.fillna(value=df['rating_count'].median())
```

```python
#   1. Best-Selling Products/Categories
category_sales = df.groupby('category')['actual_price'].sum().sort_values(ascending=False)
print("Best-selling categories:\n", category_sales.head(10))
#   Visualization
category_sales.head(10).plot(kind='bar')
plt.title('Top 10 Best-Selling Categories')
plt.show()

#   2. Impact of Discounts
df['discount_amount'] = df['actual_price'] - df['discounted_price']
sns.scatterplot(x='discount_amount', y='rating', data=df)
plt.title('Discount Amount vs. Rating')
plt.show()

#   3. Rating Analysis
print("\nAverage rating:", df['rating'].mean())
sns.histplot(df['rating'], bins=5, discrete=True)
plt.title('Distribution of Ratings')
plt.show()

#   ...   More product-focused analysis
```

```python
import pandas as pd
from textblob import TextBlob  # For basic sentiment analysis
def analyze_sentiment(text):
    if isinstance(text, str):
        analysis = TextBlob(text)
        return analysis.sentiment.polarity  # -1 to 1 (negative to positive)
    else:
        return 0  # Handle non-string cases

df['review_sentiment'] = df['review_content'].apply(analyze_sentiment)

print(df[['review_content', 'review_sentiment']].head())

#   Analyze sentiment vs. rating
sns.scatterplot(x='rating', y='review_sentiment', data=df)
plt.title('Rating vs. Review Sentiment')
plt.show()

#   ...   More review analysis (e.g., most common words)
```

```python
#   1.   Price Distribution
sns.histplot(df['actual_price'], kde=True)
plt.title('Distribution of Actual Prices')
plt.show()

#   2.   Discount vs. Rating
sns.scatterplot(x='discount_percentage', y='rating', data=df)
plt.title('Discount Percentage vs. Rating')
plt.show()

#   3.   Correlation Analysis
correlation_matrix = df[['actual_price', 'discounted_price', 'discount_percentage', 'rating', 'rating_count']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
